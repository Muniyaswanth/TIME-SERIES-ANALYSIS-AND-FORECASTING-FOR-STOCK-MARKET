# Model Implementation

## Train Test Split

```python
test_Variable=data['Close']
```

```python
train_size = int(len(test_Variable) * 0.8)
train, test = test_Variable[:train_size], test_Variable[train_size:]

print("Train size:", len(train), "Test size:", len(test))
```

## Model Building

### ARIMA Model

```python
from statsmodels.tsa.arima.model import ARIMA
model_arima = ARIMA(train, order=(5,1,0))
model_arima_fit = model_arima.fit()
forecast_arima = model_arima_fit.forecast(steps=len(test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(train, label="Train")
plt.plot(test, label="Test")
plt.plot(test.index, forecast_arima, label="ARIMA Forecast")
plt.legend()
plt.show()
```

### Model Evoluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
rmse_arima = math.sqrt(mean_squared_error(test, forecast_arima))
print("ARIMA RMSE:", rmse_arima)
```

### SARIMA Model

```python
import warnings
warnings.filterwarnings("ignore")
from statsmodels.tsa.statespace.sarimax import SARIMAX
model_sarima = SARIMAX(train, order=(2,1,2), seasonal_order=(1,1,1,12))
model_sarima_fit = model_sarima.fit(disp=False)
forecast_sarima = model_sarima_fit.forecast(steps=len(test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(train, label="Train")
plt.plot(test, label="Test")
plt.plot(test.index, forecast_sarima, label="SARIMA Forecast")
plt.legend()
plt.show()
```

### Model Evoluation

```python
rmse_sarima = math.sqrt(mean_squared_error(test, forecast_sarima))
print("SARIMA RMSE:", rmse_sarima)
```

### LSTM

```python
# Quick LSTM (requires tensorflow)
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

series = test_Variable.values.reshape(-1,1)
scaler = MinMaxScaler()
scaled = scaler.fit_transform(series)

look_back = 20
def create_seq(data, lb):
    X, y = [], []
    for i in range(lb, len(data)):
        X.append(data[i-lb:i,0])
        y.append(data[i,0])
    X = np.array(X).reshape(-1, lb, 1)
    return X, np.array(y)

train_scaled = scaled[:train_size]
test_scaled = scaled[train_size - look_back:]
X_train, y_train = create_seq(train_scaled, look_back)
X_test, y_test = create_seq(test_scaled, look_back)

model = Sequential()
model.add(LSTM(32, input_shape=(look_back,1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1)

preds = model.predict(X_test).flatten()
preds_inv = scaler.inverse_transform(preds.reshape(-1,1)).flatten()
pd.DataFrame({"actual": test.values[:len(preds_inv)], "forecast": preds_inv}, index=test.index[:len(preds_inv)])\
  .to_csv("E:/datamites/forecast_lstm_fast.csv")
print("Saved E:/datamites/forecast_lstm_fast.csv")
```

### Model Evolution

```python
min_len = min(len(y_test), len(preds_inv))
y_true = y_test[:min_len]
y_pred = preds_inv[:min_len]

mae = mean_absolute_error(preds, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

print(f"LSTM Evaluation:")
print(f"MAE  = {mae:.4f}")
print(f"RMSE = {rmse:.4f}")
print(f"MAPE = {mape:.2f}%")
```

```python
plt.figure(figsize=(12,6))
plt.plot(y_true, label='Actual')
plt.plot(y_pred, label='LSTM Forecast')
plt.legend()
plt.title("LSTM Forecast vs Actual")
plt.show()
```
